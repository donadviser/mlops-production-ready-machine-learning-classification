```python
import pandas as pd
from sklearn import datasets

from evidently.report import Report
from evidently.metric_preset import DataDriftPreset

iris_data = datasets.load_iris(as_frame=True)
iris_frame = iris_data.frame
 
```

```python
iris_frame
```

```python
data_drift_report = Report(metrics=[
    DataDriftPreset(),
])
```

```python
data_drift_report.run(current_data=iris_frame.iloc[:60], reference_data=iris_frame.iloc[60:], column_mapping=None)
data_drift_report
```

```python
data_drift_report.save_html("file.html")
```

```python
data_url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
boston_frame = pd.read_csv(data_url)
```

```python
boston_frame
```

```python
data_drift_report = Report(metrics=[
    DataDriftPreset(),
])
```

```python
data_drift_report.run(current_data=boston_frame.iloc[:150], reference_data=boston_frame.iloc[150:], column_mapping=None)
data_drift_report
```

```python
data_drift_report.save_html("boston_data_drift_report.html")
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_dataset, test_dataset = train_test_split(boston_frame, test_size=0.2, random_state=42)
```

```python
data_drift_report.run(current_data=test_dataset, reference_data=train_dataset, column_mapping=None)
data_drift_report
```

```python
data_drift_report.save_html("boston_data_drift_20slipt_report.html")
```
